#### Basic operations
---

```python
# Populate arrays 

array_1d = np.array([1.2, 2.1, 3.3, 4.1, 5.4, 6.8, 5.9, 9.8])
print('\n1d array: \n', array_1d)

array_2d = np.array([[1,0], [2,9], [8,8]])
print('\n2d array: \n',array_2d)

array_3d = np.array([ [[1,2], [2,3], [3,1]], [[1,0], [2,9], [8,8]] ])
print('\n3d array: \n',array_3d)
```

```python
# Populate with a sequence of numbers
seq_int = np.arange(1, 9)
print('\n sequence of integers: \n', seq_int)

# Populate arrays with random numbers (6 element vector + values from 50 to 100)
rand_int = np.random.randint(low = 50, high = 101, size = 6) # note high = upper limit + 1
print('\n random integers: \n', rand_int)

# Populat array with random floats from 0 to 1
rand_float = np.random.random(6)
print('\n random floats 0 to 1: \n', rand_float.round(2)) # round(n) each value to n decimal places
```

---
#### Math operations
---

If you want to add / subtract / multiply / divide two vectors or matrices, linear algebra requires that the two operands have the same dimensions. Luckily, NumPy uses a trick called broadcasting which expands the smaller operands dimensions such they are compatible for linear algebra. 

For example:

```python
rand_float_23 = rand_float + 2.0
print('Random floats from 2 to 3: \n', rand_float_23.round(2))

rand_int_x3 = rand_int * 3
print('\nRandom ints 150 to 300: \n', rand_int_x3)
```

```python
# Create a linear dataset
#    Assign a sequence of integers (n) from 8 to 88, inclusive, to an arrage named f1
#    Assign n values to an array called label, such that label = 3(f1) + 6

low = 8
high = 88
n = high - low + 1

f1 = np.sort(np.random.randint(low = low, high = high + 1, size = n))
f1_alt = np.arange(low, high + 1)

label = 3 * f1 + 6
label_alt = 3 * f1_alt + 6

print('Method 1 (randomised): \n', f1, '\n', label)
print('\nMethod 2 (sequential): \n', f1_alt, '\n', label_alt)
```

```python
# Add noise to the dataset
#    insert noise intot he label set by adding a random float (-3 / +3) to each value

noise = (np.random.random([n]) * 6) - 3 
label1 = label + noise
```

#### Meet the DataFrame (and Series data)
---

The primary structures are implemented as two classes:

* DataFrame; which you can imagine as a relational data table, with rows and named columns.
* Series; which is a single column. 

A DataFrame contains one or more Series and a name for each Series.

DataFrame objects can be created by passing a dict mapping (string) column names to their respective Series. If the Series don't match in length then missing values are NaN values

```python
council = pd.Series(['Edinburgh', 'Glasgow', 'Stirling', 'Fife'])
population = pd.Series([518500, 626400, 94330, 371910])
area = pd.Series([264, 175, 2187, 1325])

df = pd.DataFrame({'Council':council, 'Population':population, 'Area':area})
df['Density'] = round(df['Population'] / df['Area'], 1)
df 
```

#### Indexes
---

Both Series and DataFrame objects also define an index property that assigns an identifier value to each Series item or DataFrame row.

By default, at construction, pandas assigns index values that reflect the ordering of the source data. Once created, the index values are stable; that is, they do not change when data is reordered.

```python
df.index
```

```python
# We can re-index
df.reindex([3, 1, 0, 2])
```

```python
# the example below, we take the index, which is array-like, and pass it to NumPy's random.permutation function, which shuffles its values in place. Calling reindex with this shuffled array causes the DataFrame rows to be shuffled in the same way. Try running the following cell multiple times!
```

```python
df.reindex(np.random.permutation(df.index))
```
